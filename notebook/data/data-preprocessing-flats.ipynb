```python
def categorize_age_possession(value):
    if pd.isna(value):
        return "Undefined"
    if "0 to 1 Year Old" in value or "Within 6 months" in value or "Within 3 months" in value:
        return "New Property"
    if "1 to 5 Year Old" in value:
        return "Relatively New"
    if "5 to 10 Year Old" in value:
        return "Moderately Old"
    if "10+ Year Old" in value:
        return "Old Property"
    if "Under Construction" in value or "By" in value:
        return "Under Construction"
    try:
        # For entries like 'May 2024'
        int(value.split(" ")[-1])
        return "Under Construction"
    except:
        return "Undefined"
```

```python
# Extract all unique furnishings from the furnishDetails column
all_furnishings = []
for detail in df_1['furnishDetails'].dropna():
    furnishings = detail.replace('[', '').replace(']', '').replace("'", "").split(', ')
    all_furnishings.extend(furnishings)
unique_furnishings = list(set(all_furnishings))

# Define a function to extract the count of a furnishing from the furnishDetails
def get_furnishing_count(details, furnishing):
    if isinstance(details, str):
        if f"No {furnishing}" in details:
            return 0
        pattern = re.compile(f"(\d+) {furnishing}")
        match = pattern.search(details)
        if match:
            return int(match.group(1))
        elif furnishing in details:
            return 1
    return 0

# Simplify the furnishings list by removing "No" prefix and numbers
columns_to_include = [re.sub(r'No |\d+', '', furnishing).strip() for furnishing in unique_furnishings]
columns_to_include = list(set(columns_to_include))  # Get unique furnishings
columns_to_include = [furnishing for furnishing in columns_to_include if furnishing]  # Remove empty strings

# Create new columns for each unique furnishing and populate with counts
for furnishing in columns_to_include:
    df_1[furnishing] = df_1['furnishDetails'].apply(lambda x: get_furnishing_count(x, furnishing))

# Create the new dataframe with the required columns
furnishings_df = df_1[['furnishDetails'] + columns_to_include]
```

```python
furnishings_df.drop(columns=['furnishDetails'],inplace=True)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(furnishings_df)
wcss_reduced = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    wcss_reduced.append(kmeans.inertia_)
```

```python
# Plot the results
plt.figure(figsize=(12, 8))
plt.plot(range(1,11), wcss_reduced, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal Number of Clusters (Reduced Range)')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
n_clusters = 3

# Fit the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(scaled_data)

# Predict the cluster assignments for each row
cluster_assignments = kmeans.predict(scaled_data)
```

```python


# 1. Load your dataset into df_1


# 2. Create the master extraction function
def get_area(text, area_type):
    """
    Searches the text for a specific area type and extracts the first number after it.
    """
    if pd.isna(text):
        return None
        
    # Regex breakdown:
    # area_type : Looks for the exact phrase (e.g., "Carpet area")
    # \s*:?\s* : Allows for optional spaces or colons (handles "Carpet area:" vs "Carpet area ")
    # (\d+\.?\d*) : Captures the actual number, including decimals
    pattern = area_type + r'\s*:?\s*(\d+\.?\d*)'
    
    match = re.search(pattern, str(text), re.IGNORECASE)
    if match:
        return float(match.group(1))
    return None

# 3. Extract the three main area measurements into their own columns in df_1
df_1['super_built_up_area'] = df_1['areaWithType'].apply(lambda x: get_area(x, 'Super Built up area'))
df_1['built_up_area'] = df_1['areaWithType'].apply(lambda x: get_area(x, 'Built Up area'))
df_1['carpet_area'] = df_1['areaWithType'].apply(lambda x: get_area(x, 'Carpet area'))

# 4. Extract 'Plot area' (mostly found in independent houses/villas)
df_1['plot_area'] = df_1['areaWithType'].apply(lambda x: get_area(x, 'Plot area'))

# 5. Handle the Missing Data Fallback
# If a property has a 'plot_area' but no 'built_up_area', we safely assume the plot area 
# is the primary built-up area and fill those NaN values.
df_1['built_up_area'] = df_1['built_up_area'].fillna(df_1['plot_area'])

# 6. Clean up: Drop the temporary 'plot_area' and the old messy column from df_1
df_1.drop(columns=['plot_area', 'areaWithType'], inplace=True)

# 7. Check the results!
print(df_1[['super_built_up_area', 'built_up_area', 'carpet_area']].head(10))
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest

# 1. Load the data
df_2 = pd.read_csv('flats_cleanedv2.csv')

# 2. Define the exact columns you asked for
cols = ['price', 'area', 'price_per_sqft', 'bedRoom', 'luxury_score', 'furnishing_type', 'bathroom', 'balcony']

# 3. Clean 'balcony' column if it contains strings like '3+'
if df_2['balcony'].dtype == object:
    df_2['balcony'] = df_2['balcony'].astype(str).str.replace('+', '', regex=False).str.replace('No', '0', regex=False).astype(float)

# 4. Drop NaNs so the model can run perfectly
df_2_clean = df_2.dropna(subset=cols).copy()

# 5. Initialize Isolation Forest (assuming 5% of your data are outliers)
iso = IsolationForest(contamination=0.05, random_state=42)

# 6. Predict outliers (-1 means Outlier, 1 means Normal)
df_2_clean['outlier'] = iso.fit_predict(df_2_clean[cols])

# 7. Extract and view the anomalies!
anomalies = df_2_clean[df_2_clean['outlier'] == -1]

print(f"Total Outliers Found: {len(anomalies)}")

# Show the top outliers sorted by highest price and largest area
top_outliers = anomalies[cols].sort_values(by=['price', 'area'], ascending=[False, False])
print(top_outliers.head(10))
```

```python
import pandas as pd

# Load the data
df_2 = pd.read_csv('flats_cleanedv2.csv')

# --- Our Previous Rules ---
rule_1_mask = df_2['price_per_sqft'] > 100000
rule_2_mask = (df_2['area'] / df_2['bedRoom']) < 150
rule_3_mask = (df_2['area'] > 10000) & (df_2['price'] < 2.0)

# --- NEW RULE: The "Too Cheap" / "Typo Area" Rule ---
# Any flat costing less than ₹2,000 per sqft is almost certainly fake or a rental listing
rule_4_mask = df_2['price_per_sqft'] < 2000

# Combine all masks
fake_data_mask = rule_1_mask | rule_2_mask | rule_3_mask | rule_4_mask

# Keep only the valid rows
df_2_clean = df_2[~fake_data_mask].copy()

print(f"Original properties: {df_2.shape[0]}")
print(f"Fake properties dropped: {fake_data_mask.sum()}")
print(f"Final, pristine dataset: {df_2_clean.shape[0]}")

# Save the final pristine dataset
# df_2_clean.to_csv('flats_ready_for_model.csv', index=False)
```

```python
import pandas as pd
import re

# Load the dataset
df_3 = df_2_clean.copy()

def extract_sector_generalized(value):
    if not isinstance(value, str):
        return value
    
    # Regex pattern to find 'Sector' and extract ONLY the digits that follow it
    match = re.search(r"(?i)Sector[\s-]*(\d+)", value)
    
    if match:
        # Group 1 captures only the numbers (e.g., '88' from 'Sector 88B')
        return match.group(1) 
    else:
        # Keep original value if 'Sector' pattern is not found (e.g., 'Krishna Colony')
        return value

# Apply the generalized extraction logic
df_3['sector'] = df_3['sector'].apply(extract_sector_generalized)

# Save the result
df_3.to_csv('flats_cleaned_v4.csv', index=False)
```
